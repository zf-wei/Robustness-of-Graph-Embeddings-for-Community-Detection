```python
from WGE.graph_input import graph_input_simple
import numpy as np
import networkx as nx
```

```python
N=23748
mu=0.01
disturb_type=1

graph, membership, between = graph_input_simple(N=N, mu=mu, disturb_type=disturb_type)  

sorted_betweenness = sorted(between)
ranking = {value: index + 1 for index, value in enumerate(sorted_betweenness)}
ranking_weight =  [ranking[number] for number in between]
        
output_file = f"graph_{N}_{mu}.rank"
np.savetxt(output_file, ranking_weight, fmt='%d', delimiter=' ')
```

```python
# Get the number of connected components
num_connected_components = nx.number_connected_components(graph)

# Print the number of connected components
print("Number of Connected Components:", num_connected_components)
```
